## Список наименований средств, соответствующих двум любым документам с требованиями по сертификации

В параметре "Наименования документов, требованиям которых соответствует средство" перечислены различные документы. Нашей задачей является найти те средства, которые представлены двумя документами в этом параметре. Эти документы перечислены через запятую, поэтому для нахождения, будем определять кличество запятых в строке.

```python
def find_two_docs(str):
    """Функция нахождения количества запятых в строке"""
    
    p = re.compile(',')
    
    # Количество запятых в строке  
    coma_len = len(p.findall(str))
    
    return coma_len
```

```python
# Примянем функцию к каждой строке датасета
coma_len = df['Наименования документов, требованиям которых соответствует средство'].apply(find_two_docs)
```

```python
df['coma_len'] = coma_len
df_two_docs = df[df['coma_len'] == 1]
```

```python
# Список наименований средств, соответствующих двум любым документам с требованиями по сертификации
df_two_docs['Наименование средства (шифр)']
```

## Наименования средств, которые имеют просроченные сертификаты

```python
# Удаление пустых значений
df_date_term = df.dropna(subset = 'Срок действия сертификата')
str_date_term = df_date_term[df_date_term['Срок действия сертификата']=='#Н/Д'].index
df_date_term.drop(str_date_term, axis=0, inplace=True)
```

```python
# Перевод в формат даты "Срок действия сертификата"
df_date_term['Срок действия сертификата'] \
     = pd.to_datetime(df_date_term['Срок действия сертификата'])
```

```python
def return_overdue():
    """Средства, которые имеют просроченные сертификаты"""
    
    # Дата актуального дня
    date_today = pd.to_datetime(datetime.date.today())
    
    # Датасет просроченных сертификатов
    df_overdue = df_date_term[df_date_term['Срок действия сертификата'] < date_today]
    
    return df_overdue
```

```python
# Наименования средств, которые имеют просроченные сертификаты
funds_overdue = return_overdue()['Наименование средства (шифр)']
funds_overdue
```

## Наименования средств, которые имеют просроченные сертификаты, но при этом обязаны оказывать поддержку

```python
# Удаление пустых значений 
df_date = return_overdue().dropna(subset = 'Информация об окончании срока технической поддержки, полученная от заявителя')

# Наименовая средства, у которых бессрочная поддержка выделяем в отдельную переменную 
str_date = df_date[df_date['Информация об окончании срока технической поддержки, полученная от заявителя']=='бессрочно']
df_date.drop(str_date.index, axis=0, inplace=True)
```

```python
# Перевод в формат даты
df_date['Информация об окончании срока технической поддержки, полученная от заявителя'] \
     = pd.to_datetime(df_date['Информация об окончании срока технической поддержки, полученная от заявителя'])
```

```python
def return_overdue_and_support():
    """Средства, которые имеют просроченные сертификаты, но должны оказывать поддержку"""
    
    # Дата актуального дня
    date_today = pd.to_datetime(datetime.date.today())
    
    # Датасет средств, которые обязаны оказывать поддержку на аткульный день
    df_overdue = df_date[df_date['Информация об окончании срока технической поддержки, полученная от заявителя'] > date_today]
    
    return df_overdue
```

```python
funds_overdue_and_support = return_overdue_and_support()

#funds_overdue_and_support.concat()

# Наименования средств, которые имеют просроченные сертификаты, но при этом обязаны оказывать поддержку
funds_overdue_and_support['Наименование средства (шифр)']
```

## Наименования всех средств, которые проходили испытания в частных организациях (форма организации ООО, ЗАО, ПАО, АО).

```python
def find_private_organiz(str):
    """Поиск форм организации в строке"""
    
    # Возращает 0, если в строке есть хотя бы один из представленных символов
    private_organiz_exist = str.find("ООО ") * str.find("ЗАО ") * str.find("ПАО ") * str.find("АО ")
    
    return private_organiz_exist
```

```python
# Применяем функцию к каждоой строке параметра 'Испытательная лаборатория'
df['find_private_organiz'] = df['Испытательная лаборатория'].apply(find_private_organiz)

# Датасет средств, которые проходили испытания в частных организациях
df_private_org = df[df['find_private_organiz'] == 0]
```

```python
# Наименования всех средств, которые проходили испытания в частных организациях
df_private_org['Наименование средства (шифр)']
```
